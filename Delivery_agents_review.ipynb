```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in ['Agent Name', 'Location', 'Order Type','Customer Feedback Type','Price Range','Discount Applied','Product Availability','Order Accuracy']:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
df.drop('Review Text', axis=1, inplace=True)
```

```python
(df['Discount Applied'].value_counts(normalize=True) * 100)
# What percentage of orders had discounts applied?
```

```python
df.groupby('Order Type')['Delivery Time (min)'].mean()
# average delivery time per order type
```

```python
df.groupby('Location')['Rating'].mean()
# average rating for agents in each location
```

```python
df.groupby('Product Availability')['Rating'].mean()
# does product availability affect average rating
```

```python
df.groupby('Price Range')['Customer Service Rating'].mean().sort_values(ascending=False)
# Which price range has the highest average customer service rating?
```

```python
# Prepare the data
average_rating = df['Rating'].mean()
average_delivery_time = df['Delivery Time (min)'].mean()
top_locations = df['Location'].value_counts().head(5)
order_types = df['Order Type'].value_counts()
csr_distribution = df['Customer Service Rating'].value_counts().sort_index()
discount_percentage = df['Discount Applied'].value_counts(normalize=True) * 100
delivery_time_by_type = df.groupby('Order Type')['Delivery Time (min)'].mean()
rating_by_location = df.groupby('Location')['Rating'].mean()
rating_by_availability = df.groupby('Product Availability')['Rating'].mean()
csr_by_price_range = df.groupby('Price Range')['Customer Service Rating'].mean().sort_values(ascending=False)

# Print results
print("Average Rating:", average_rating)
print("Average Delivery Time:", average_delivery_time)
print("Top Locations:\n", top_locations)
print("Order Types:\n", order_types)
print("Customer Service Rating Distribution:\n", csr_distribution)
print("Discount Applied (%):\n", discount_percentage)
print("Avg Delivery Time by Order Type:\n", delivery_time_by_type)
print("Avg Rating by Location:\n", rating_by_location)
print("Avg Rating by Product Availability:\n", rating_by_availability)
print("Customer Service Rating by Price Range:\n", csr_by_price_range)
```
